### Grouping Category and Summing the Score Value

```python
grouped_df = new_df.groupby('Category')['Score'].sum().reset_index()
grouped_df
```

### Calculating Top 5 Category Based on Score

```python
top_5 = grouped_df.sort_values(by=["Score"],ascending=False).head(5)
top_5
```

### Calculating Bottom 5 Category Based on Score

```python
bot_5 = grouped_df.sort_values(by=["Score"],ascending=True).head(5)
bot_5
```

### Percentage of each Category

```python
total_score = top_5['Score'].sum()
top_5['Percentage'] = ((top_5['Score'] / total_score) * 100).round(2)
top_5
```

### Visualization

```python
x = top_5["Category"]
y = top_5["Score"]

plt.bar(x,y)
plt.xticks(rotation = 90)
plt.ylim(60000, 76000)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_5['Percentage'], labels=top_5['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Category Percentages')
plt.axis('equal')
```
